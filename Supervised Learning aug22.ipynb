```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
#from imblearn.combine import SMOTEENN
```

```python
train_data=pd.read_csv("C:/Users/Waleed/Downloads/churn-bigml-80.csv")
train_data.head()
```

```python
test_data=pd.read_csv("C:/Users/Waleed/Downloads/churn-bigml-20.csv")
test_data.head()
```

```python
train_data.info()
```

```python
test_data.info()
```

```python
train_data['Churn'] = train_data['Churn'].map({ False: 0, True: 1 })
#train_data['International plan'] = train_data['International plan'].map({ 'No': 0, 'Yes': 1 })
#train_data['Voice mail plan'] = train_data['Voice mail plan'].map({ 'No': 0, 'Yes': 1 })
train_data=train_data.drop("State",axis=1)
train_data=train_data.drop("International plan",axis=1)
train_data=train_data.drop("Voice mail plan",axis=1)
```

```python
test_data['Churn'] = test_data['Churn'].map({ False: 0, True: 1 })
#train_data['International plan'] = train_data['International plan'].map({ 'No': 0, 'Yes': 1 })
#train_data['Voice mail plan'] = train_data['Voice mail plan'].map({ 'No': 0, 'Yes': 1 })
test_data=test_data.drop("State",axis=1)
test_data=test_data.drop("International plan",axis=1)
test_data=test_data.drop("Voice mail plan",axis=1)
```

```python
train_data.info()
```

```python
y_train=train_data["Churn"]
y_train.head()
```

```python
y_test=test_data["Churn"]
y_test.head()
```

```python
x_train=train_data.drop("Churn",axis=1)
x_train.head()
```

```python
x_test=test_data.drop("Churn",axis=1)
x_test.head()
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

# Decision Tree Model

```python
dt=DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
y_pred
```

```python
dt.score(x_test,y_test)
```

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```

# Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=100, criterion='gini', random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
y_pred
```

```python
rf.score(x_test,y_test)
```
